```python

# Load scikit's random forest classifier library
from sklearn.ensemble import RandomForestClassifier
#SVC
from sklearn.svm import SVC, LinearSVC
#KNN
from sklearn.neighbors import KNeighborsClassifier
#perceptron
from sklearn.linear_model import Perceptron
# Load pandas
import pandas as pd
from pandas import DataFrame
# Load numpy
import numpy as np
#df=pd.read_csv('eve_2.csv')
df=pd.read_csv('prod_data.csv')
df.head()
```

```python
#df[' payment_response-recieved'].value_counts()
df['payment_received'].value_counts()
```

```python
#User_id=df.User_Id
user_id = df.user_id
df=df.drop('user_id',1)
```

```python
#y=df[' payment_response-recieved']
y=df['payment_received']
```

```python
#X=df.drop(' payment_response-recieved',1)
X=df.drop('payment_received',1)
```

```python
clf = RandomForestClassifier(n_estimators=100)
```

```python
X=X.fillna(0)
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
clf.fit(X_train,y_train)
print(X_test)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
probab=clf.predict_proba(X_test)
```
